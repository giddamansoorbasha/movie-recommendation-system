## **Loading Data**

```python
import pandas as pd
```

```python
movies = pd.read_csv('../data/tmdb_5000_movies.csv')
credits = pd.read_csv('../data/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

## **Merging Data**

```python
movies = movies.merge(credits, on='title')
```

```python
movies.shape
```

```python
movies.head(1)
```

## **Selected Cols**

```python
movies = movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
movies.shape
```

## **Dropping Nulls**

```python
movies.isnull().sum()
```

```python
movies = movies.dropna()
```

```python
movies.shape
```

## **Adding Movie Image Paths**

```python
poster_path = pd.read_csv('../data/poster_path.csv')
```

```python
poster_path.head()
```

```python
movies = movies.merge(poster_path[['id', 'poster_path']], on='id', how='left')
```

```python
movies.head()
```

```python
movies.to_csv('../data/movies.csv')
```
